```python
import copy
import numpy as np
from collections import Iterable
from scipy.stats import truncnorm
import torch
import torch.nn as nn
from adversarialbox.utils import to_var
from sampler.sghmc import SGHMC
class BayesWRM(object):
    def __init__(self, model=None, model_list=None, Stheta=2, Sz=2, epsilon=0.5, k=15, a=1, gamma=2, T=1e-3, random_start=False):
        """
        Attack parameter initialization. The attack performs k steps of
        size a, while always staying within epsilon from the initial
        point.
        https://github.com/MadryLab/mnist_challenge/blob/master/pgd_attack.py
        """
        self.model = model
        self.epsilon = epsilon
        self.k = k
        self.a = a
        self.rand = random_start
        self.gamma = gamma
        self.loss_fn = nn.CrossEntropyLoss()
        
        self.T = T #Temperature
        self.model_list = model_list
        
        self.Stheta = Stheta
        self.Sz = Sz

    def perturb(self, X_nat, y):
        """
        Given examples (X_nat, y), returns adversarial
        examples within epsilon of X_nat in l_infinity norm.
        """
        if self.rand:
            X = X_nat + np.random.uniform(-self.epsilon, self.epsilon,
                X_nat.shape).astype('float32')
        else:
            X = np.copy(X_nat)
        Sz = self.Sz
        Stheta = self.Stheta
        MC = 15
        #List of numpy arrays
        z_list = [copy.deepcopy(X) for i in range(Sz)]
        
        
        y_var = to_var(torch.LongTensor(y))
        loss_fn = nn.CrossEntropyLoss()
        for k in range(Sz):
            
            class ZNN(nn.Module):
                def __init__(self):
                    super(ZNN, self).__init__()
                    self.Z_var = nn.Parameter(torch.from_numpy(z_list[k].astype(np.float32)).cuda(), requires_grad=True)
                    
                def forward(self, model_list, gamma):
                    
                    Z_var0 = Variable(torch.from_numpy(z_list[k].astype(np.float32)).cuda(), requires_grad=False)
                    total_loss = 0
                    for i in range(Stheta):
                        scores = model_list[i](self.Z_var)
                        #loss = self.loss_fn(scores, y_var) - self.gamma*torch.sum((self.Z_var-Z_var0)**2)
                        #Tracer()()
                        loss = loss_fn(scores, y_var) - gamma*torch.sum((self.Z_var-Z_var0)**2)
                    total_loss = total_loss + loss
                    return total_loss
                
            znn = ZNN()         
            #optimizer = torch.optim.SGD(znn.parameters(), lr=1e-4)  
            optimizer = SGHMC(znn.parameters(), config=dict(lr=1e-4))
            
            def helper():
                def feval():
                    total_loss = znn(model_list=self.model_list, gamma=self.gamma)
                    optimizer.zero_grad()
                    total_loss.backward()
                    return total_loss #TODO return loss for extension
                return feval
            total_loss = optimizer.step(helper())
            z_list[k] = znn.Z_var.data.cpu().numpy()

        
        print('Sz step finished')
        return z_list
        
        print('maximum diff adv', np.max(np.abs(X-X_nat)))
```
