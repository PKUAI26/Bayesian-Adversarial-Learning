```python
import copy
import numpy as np
from collections import Iterable
from scipy.stats import truncnorm
import torch
import torch.nn as nn
from adversarialbox.utils import to_var

class BayesWRM(object):
    def __init__(self, model=None, model_list=None, epsilon=0.5, k=15, a=1, gamma=2, T=1e-3, random_start=False):
        """
        Attack parameter initialization. The attack performs k steps of
        size a, while always staying within epsilon from the initial
        point.
        https://github.com/MadryLab/mnist_challenge/blob/master/pgd_attack.py
        """
        self.model = model
        self.epsilon = epsilon
        self.k = k
        self.a = a
        self.rand = random_start
        self.gamma = gamma
        self.loss_fn = nn.CrossEntropyLoss()
        
        self.T = T #Temperature
        self.model_list = model_list

    def perturb(self, X_nat, y):
        """
        Given examples (X_nat, y), returns adversarial
        examples within epsilon of X_nat in l_infinity norm.
        """
        if self.rand:
            X = X_nat + np.random.uniform(-self.epsilon, self.epsilon,
                X_nat.shape).astype('float32')
        else:
            X = np.copy(X_nat)
        Sz = 2
        Stheta = 2
        MC = 15
        #List of numpy arrays
        z_list = [copy.deepcopy(X) for i in range(Sz)]
        
        
        y_var = to_var(torch.LongTensor(y))
        for k in range(Sz):
            #Langevin MC steps
            for m in range(MC):
                #Aggregate loss gradients
                Z_var = to_var(torch.from_numpy(z_list[k].astype(np.float32)), requires_grad=True)
                Z_var0 = deepcopy(Z_var)
                #Z_var = Z_var.type(torch.cuda.FloatTensor)
                #Tracer()()
                total_loss = 0
                for i in range(Stheta):
                    scores = self.model_list[i](Z_var)
                    loss = self.loss_fn(scores, y_var) - self.gamma*torch.sum((Z_var-Z_var0)**2)
                    total_loss = total_loss + loss
                
                
                total_loss.backward()
                grad_L_Z = Z_var.grad.data.cpu().numpy()
                self.a = self.a/np.sqrt(m+1)
                z_list[k] = z_list[k] + self.a * grad_L_Z + np.sqrt(2*self.a*self.T)*np.random.normal(size=z_list[k].shape)   
                
        print('Sz step finished')
        return z_list
        
        print('maximum diff adv', np.max(np.abs(X-X_nat)))
```

```python
from torch.optim import Optimizer
from copy import deepcopy
#TODO ref to entropy sgd to implement SGHMC, iteration steps inside the class
class SGHMC(Optimizer):
    def __init__(self, params, lr=0, alpha=0):
        defaults = dict(lr=lr, alpha=alpha)
        super(SGHMC, self).__init__(params, defaults)
    
    def __setstate__(self, state):
        super(SGHMC, self).__setstate__(state)
    
    def step(self, closure=None):
        loss = None
        if closure is not None:
            loss = closure()
            
          
        #parameter setting
        L = int(15) #number of Langevin steps
        gamma = 0.01
        lr=0.5
        T = 1e-5
        #inner sampling steps
        for i in range(L):
            lr = lr/np.sqrt(L+1)
            f = closure()
            
            for group in self.param_groups:
                for param in group['params']:
                    if param.grad is None:
                        continue
                    d_param = param.grad.data
                    
                    param_state = self.state[param]
                    if 'mometum' not in param_state:
                        momentum = param_state['momentum']=torch.zeros(param.data.size())
                    
                    momentum = (1-gamma)*momentum - lr*d_param + np.sqrt(2*lr*gamma*T)*momentum.normal_()
                    param.data.add_(momentum)    
                    #momentum.mul_(1-gamma).add_(lr*d_param)
                    #Tracer()()
                    #momentum.add_(np.sqrt(2*lr*gamma)*momentum.normal_())
        return loss
```

```python
from torch.optim import Optimizer
#TODO ref to entropy sgd to implement SGHMC, iteration steps inside the class
class SGHMC(Optimizer):
    def __init__(self, params, lr=0, momentum=0):
        defaults = dict(lr=lr, momentum=momentum)
        super(SGHMC, self).__init__(params, defaults)
    
    def __setstate__(self, state):
        super(SGHMC, self).__setstate__(state)
    
    def step(self, closure=None):
        loss = None
        if closure is not None:
            loss = closure()
        
        for group in self.param_groups:
            momentum = group['momentum']
            
            for p in group['params']:
                if p.grad is None:
                    continue
                d_p = p.grad.data
                
                param_state = self.state[p]
                
                if 'momentum_buffer' not in param_state:
                    buf = param_state['momentum_buffer'] = d_p.clone()
                else:
                    buf = param_state['momentum_buffer']
                    buf.mul_(momentum).add_(1, d_p)
                #Tracer()()
                p.data.add_(-group['lr'], d_p)
        return loss
```
